# 포리움 지도 시각화

```python
import folium
folium.__version__
```

```python
map=folium.Map(location=[35.159841,126.851784],zoom_start=12)
folium.Marker(
    location=[35.159841,126.851784],
    popup=folium.Popup('광주광역시 서구 내방로 111',max_width=200),
    tooltip='광주광역시청'
).add_to(map)
folium.Marker(
    location=[35.146107, 126.92331],
    popup=folium.Popup('광주광역시 동구 서남로 1',max_width=200),
    tooltip='광주광역시 동구청'
).add_to(map)
folium.Marker(
    location=[35.15208, 126.88985],
    popup=folium.Popup('광주광역시 서구 농성동 경열로 33',max_width=200),
    tooltip='광주광역시 서구청'
).add_to(map)
folium.Marker(
    location=[35.1741004, 126.9121570],
    popup=folium.Popup('광주광역시 북구 우치로 77',max_width=200),
    tooltip='광주광역시 북구청'
).add_to(map)
folium.Marker(
    location=[35.1329009, 126.9025733],
    popup=folium.Popup('광주광역시 남구 봉선로 1',max_width=200),
    tooltip='광주광역시 남구청'
).add_to(map)
folium.Marker(
    location=[35.1395519, 126.7938068],
    popup=folium.Popup('광주광역시 광산구 광산로29번길 15',max_width=200),
    tooltip='광주광역시 광산구청'
).add_to(map)
map
```

```python
map.save('광주시_주요관공서.html')
```
